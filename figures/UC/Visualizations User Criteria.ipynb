```python
import numpy as np
import matplotlib.pyplot as plt

# Data extracted from the LaTeX table
labels = [
    "GPT-3.5\nNone", "GPT-3.5\nRAG", "GPT-3.5\nAll-in-prompt",
    "o4-mini\nNone", "o4-mini\nRAG", "o4-mini\nAll-in-prompt",
    "LearnLM\nNone", "LearnLM\nRAG", "LearnLM\nAll-in-prompt",
    "o4-mini\nNone\n(Code exec)", "o4-mini\nRAG\n(Code exec)", "o4-mini\nAll-in-prompt\n(Code exec)",
    "LearnLM\nNone\n(Code exec)", "LearnLM\nRAG\n(Code exec)", "LearnLM\nAll-in-prompt\n(Code exec)"
]
means = [
    44.05, 46.43, 60.32,
    93.26, 95.64, 95.64,
    92.07, 89.69, 88.89,
    93.26, 96.43, 94.84,
    93.65, 91.67, 92.86
]
stds = [
    2.57, 1.68, 2.03,
    1.12, 1.12, 0.56,
    0.56, 0.56, 0.56,
    0.56, 0.00, 1.12,
    0.56, 0.97, 0.97
]

# Define colors for each bar
colors = []
for i in range(len(labels)):
    if i == 0:  # 1st column
        colors.append('red')
    elif i in [4, 5, 10]:  # 5th, 6th, and 11th columns
        colors.append('green')
    else:
        colors.append('#4f98ca')

x = np.arange(len(labels))
plt.figure(figsize=(14, 6))
plt.bar(x, means, yerr=stds, capsize=5, color=colors, edgecolor='black')
plt.xticks(x, labels, rotation=45, ha='right')
plt.ylabel('Average Accuracy')
plt.title('Average Accuracy of Configurations on True/False Questions')
plt.tight_layout()

# Plot horizontal lines for the top 3 scores
top3_indices = np.argsort(means)[-3:]
for idx in top3_indices:
    plt.axhline(means[idx], color='green', linestyle='--', linewidth=1, alpha=0.7, label='Top 3' if idx == top3_indices[0] else "")

# Plot horizontal line for the lowest score
lowest_idx = np.argmin(means)
plt.axhline(means[lowest_idx], color='red', linestyle='--', linewidth=1, alpha=0.7, label='Lowest')

# Add legend (avoid duplicate labels)
handles, labels_ = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels_, handles))
plt.legend(by_label.values(), by_label.keys())

plt.show()
```
