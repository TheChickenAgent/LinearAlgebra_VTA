```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['1', '2', '3', '4', '5']
#labels = ["Not at all helpful", "Slightly helpful", "Moderately helpful", "Very helpful", "Extremely helpful"]
scores = [0, 5, 12, 24, 5]  # Example counts for each category
# Prepare data for histogram
data = [i for i, count in enumerate(scores) for _ in range(count)]
bins = np.arange(len(scores) + 1)

plt.hist(
    data,
    bins=bins,
    color='#4f98ca',
    edgecolor='black',
    rwidth=0.8
)

# Set xticks at the center of each bar
plt.xticks(np.arange(len(labels)) + 0.5, labels)
plt.xlabel('Helpfulness')
plt.ylabel('Count')
plt.title('Q1: Helpfulness Scores of the VTA')
plt.tight_layout()
plt.show()
```
