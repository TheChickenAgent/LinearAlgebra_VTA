```python
import matplotlib.pyplot as plt

# Example helpfulness scores for 5 stages (replace with your actual data if needed)
labels = ['Not at all helpful', 'Slightly Helpful', 'Moderatly Helpful', 'Very Helpful', 'Extremely Helpful']
scores = [0, 5, 12, 24, 5]  # Example counts for each category

colors = ['#ff9999','#ffc000','#8fd9b6','#4f98ca','#d395d0']

#plt.figure(figsize=(7, 7))
plt.pie(scores, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Q1: Helpfulness of the Assistant vs Course Material')
plt.axis('equal')
plt.show()
```
